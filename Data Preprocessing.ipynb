```python
#Import the diabetes dataset from sklearn.
from sklearn.datasets import load_diabetes
df = load_diabetes()
print(df.DESCR)
```

```python
#Data keys
df.keys()
```

```python
#Create dataframe for independent variable.
data = pd.DataFrame(data = df.data, columns = df.feature_names)
data
```

```python
#Create dataframe for target variable.
data_1 = pd.DataFrame(data = df.target)
data_1.columns = ["Target"]
data_1
```

```python
#Concat independent variable and Target variable.
data = pd.concat([data,data_1], axis = 1)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
#Check for the null values in the dataset.
data.isna().sum()
```

```python
#Create box plot for the attributes and target variable.
for col in data.columns:
    plt.boxplot(data[col])
    plt.title("Boxplot of column {}.".format(col))
    plt.show()
```

```python
#Treat outlier on bmi, s1,s2, s3,s4, s5,s6 columns.
Q1 = data['bmi'].quantile(0.25)
Q3 = data['bmi'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= bmi <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s1'].quantile(0.25)
Q3 = data['s1'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s1 <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s2'].quantile(0.25)
Q3 = data['s2'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s2 <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s3'].quantile(0.25)
Q3 = data['s3'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s3 <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s4'].quantile(0.25)
Q3 = data['s4'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s4 <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s5'].quantile(0.25)
Q3 = data['s5'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s5 <=(@Q3 + 1.5 * @IQR)')
Q1 = data['s6'].quantile(0.25)
Q3 = data['s6'].quantile(0.75)
IQR = Q3 - Q1
data = data.query('(@Q1 - 1.5 * @IQR) <= s6 <=(@Q3 + 1.5 * @IQR)')
```
